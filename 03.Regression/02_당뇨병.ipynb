# 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

- 당뇨병 데이터 셋

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
# 다양한 통계량을 요약
df.describe()
```

- 당뇨병(target) vs.m BMI

```python
# X(BMI), Y(당뇨병) 만들기
X = df.bmi.values.reshape(-1,1)
Y = diabetes.target
X.shape, Y.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, Y)
```

```python
weight, bias = lr.coef_, lr.intercept_
rss = lr._residues
weight[0], bias, rss
```

```python
# 잔차 제곱
lr.score(X, Y)
```

- train/tesr data 분리
- test_size = 0.1

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.1, random_state=2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
pred[:4]
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mse, rmse
```

- 그래프로 확인

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.show()
```

```python
w, b = lr.coef_, lr.intercept_
plt.scatter(X_train, y_train)
plt.plot([-0.1, 0.18], [-0.1*w+b, 0.18*w+b], 'r', lw=3)
plt.grid()
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot([-0.1, 0.18], [-0.1*w+b, 0.18*w+b], 'r', lw=3)
plt.grid()
plt.show()
```

## 회귀 표본 만들어 학습하기

```python
from sklearn.datasets import make_regression
X, y = make_regression(
    n_samples=100, n_features=1, n_targets=1, random_state=2021
    )
```

```python
plt.scatter(X, y)   # 데이터가 어떻게 생긴지 확인
plt.grid()
plt.show()  # 정규분포의 데이터 확인인
```

```python
X, y = make_regression(
    n_samples=100, n_features=1, n_targets=1, 
    noise=100, bias=10, random_state=2021
    ) # noise를 주면 임의로 데이터를 꺼내준다
plt.scatter(X, y)
plt.grid()
plt.show()  # 데이터가 확장된것을 볼 수 있다
```

```python
lr = LinearRegression()
lr.fit(X, y)
w, b = lr.coef_, lr.intercept_
```

```python
plt.scatter(X, y)
plt.plot([-3,4],[-3*w+b, 4*w+b], 'r', lw=3)
plt.grid()
plt.show()
```

```python
r2 = lr.score(X,y)
w[0], b, r2
```
